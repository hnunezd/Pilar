```python
import plot_likert
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from collections import Counter
import seaborn as sns

warnings.filterwarnings("ignore")
```

```python
tables=pd.read_excel(r"C:\Users\henm7\OneDrive\UCM\Publicaciones\Pilar\pilar.xlsx") # tienen que colocar entre comillas la ruta donde tienen el archivo
```

```python
tables.info()
```

```python
fig1 = tables.iloc[:,[3,4,5,6,7,8,9,10]]
fig1.head()
```

```python
myscale1 = \
    [1,
 2,
 3,
 4,
 5]
```

```python
ax=plot_likert.plot_likert(fig1,myscale1, bar_labels=True,plot_percentage=True,colors=plot_likert.colors.likert5,fontsize=9)
ax.figure.set_size_inches(18, 8)
ax.xaxis.set_label_text('Porcentaje de respuestas');

ax.legend(ncol=len(myscale1), bbox_to_anchor=(0.5, 1.02),
              loc="center", fontsize='small')
#plt.savefig("/content/drive/MyDrive/Graficos/Matias/grafico1.pdf"); #acá deben modificar "/content/drive/MyDrive/Graficos/Matias/" pueden colocar la misma donde tienen el archivo excel
plt.show()
```
